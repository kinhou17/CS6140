```python
import pandas as pd
from sklearn.datasets import load_boston
boston = load_boston()
housing = pd.DataFrame(boston.data)
housing.columns = boston.feature_names
housing['MEDV'] = pd.DataFrame(boston.target)
housing.head()
```

```python
# visualize the data
import matplotlib.pyplot as plt
housing.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
housing.isnull().sum(axis = 0)
```

```python
# NORMALIZATION

from sklearn.preprocessing import minmax_scale

def apply_minmax_scale(dataset, features):
    for feature in features:
        dataset[feature] = minmax_scale(dataset[[feature]])
        
numerical_features = ["CRIM","ZN", "INDUS", "CHAS", "NOX",
                      "RM", "AGE", "DIS", "RAD", 
                      "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
apply_minmax_scale(housing, numerical_features)
housing.columns
```

```python
from sklearn.model_selection import train_test_split

# y = housing["MEDV"]
# X = housing.drop(columns = "MEDV")
train, test = train_test_split(housing, test_size=0.2, random_state = 93)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
    
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# Testing against the training set

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression


y_train = train["MEDV"]
X_train = train.drop(columns = "MEDV")
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

scores = cross_val_score(lin_reg, X_train, y_train,
                         scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-scores)
display_scores(lin_rmse_scores)
```

```python
# Testing against the test set
y_test = test["MEDV"]
X_test = test.drop(columns = "MEDV")
scores = cross_val_score(lin_reg, X_test, y_test,
                         scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-scores)
display_scores(lin_rmse_scores)
```

```python
# Trying concatenating RM and LSTAT (for collinearity) to see if we get better score
X = pd.DataFrame(np.c_[housing['LSTAT'], housing['RM']], columns = ['LSTAT','RM'])
y = housing['MEDV']
lin_reg = LinearRegression()
lin_reg.fit(X, y)
scores = cross_val_score(lin_reg, X_test, y_test,
                         scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-scores)
display_scores(lin_rmse_scores)
```
